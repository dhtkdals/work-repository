# 크롤링 한 사이트Transfermarkt

#  이 사이트를 크롤링 한 이유 : 평소에 축구에 관심이 있어 축구 선수 시장 가치 사이트인 transfermarkt을 크롤링하였다 선수의 특징 정보를 수집하고 싶었다.


requests: 웹 페이지에서 정보를 요청하기 위한 라이브러리
BeautifulSoup: HTML 문서를 파싱하고 원하는 데이터를 추출하기 위한 라이브러리
pandas: 데이터를 효율적으로 다루기 위한 라이브러리
time: 일정한 시간 동안 프로그램을 일시 정지하기 위한 라이브러리

number, name, position, age, nation, team, value 리스트들을 생성하여 선수 정보를 저장할 예정입니다.

반복문을 통해 두 페이지를 가져온다.
각 페이지의 URL을 생성하고, 해당 URL에 요청을 보내서 페이지 내용을 가져온다.
BeautifulSoup을 사용하여 페이지 내용을 파싱한다.
선수 정보가 담긴 HTML 테이블 행을 찾아서 필요한 정보를 리스트에 추가한다.
각 페이지를 스크래핑한 후 1초 동안 대기한다.
모든 페이지를 스크래핑한 후, 선수 정보를 Pandas DataFrame으로 저장한다.
사용자가 원하는 선수의 정보를 검색하기 위해 find_player_info 함수를 정의한다.
입력받은 선수명으로 DataFrame에서 해당 선수 정보를 찾아 반환한다.
무한 루프를 통해 사용자에게 선수명을 입력받는다.
입력한 선수명으로 정보를 검색하고, '종료'를 입력하면 루프가 종료된다.
